```python
#Python 1
from collections import Counter

def fun(sen:str)->int:
  try:
    sen_list=sen.split(' ')
    dic=Counter(sen_list)
    max_value=max(dic.values())
    list_max_word=[]
    for i in dic.keys():
      if dic[i]==max_value:
        list_max_word.append(i)
    max_val=0
    for word in list_max_word:
      if len(word)>max_val:
        max_val=len(word)
    return max_val
  except:
    print('error')
 
fun('write write write all the number from from from 1 to 100')
```

```python
# python 2
from collections import Counter
def valid_or_not(st:str)->str:
  d=Counter(st)
  if len(set(d.values()))==1:
    return 'yes'
  elif len(set(d.values()))==2:
    a=Counter(d.values())
    if 1 in d.values():
      return 'yes'
    elif a[max(d.values())]==1 and max(d.values())-min(d.values()):
      return 'yes'
    else:
      return 'no'
  else:
    return 'no'
valid_or_not('abbbccc')

    
```

```python
# python 3
import pandas as pd
from json import loads,dumps
def read_json(link,to_file):
  try:
    df=pd.read_json(link)
    j=df.to_json()
    j=loads(j)
    df=pd.DataFrame.from_records(j['pokemon']).T.reset_index()
    df.drop(columns=['index'],inplace=True)
    df.to_csv(f'{to_file}.csv',index=False)
    return f'save to {to_file}.csv'
  except:
    print('error')

read_json('https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json','data')
```

```python
# python 4
def nasa_json_data(link,to_file):
  try:
    df=pd.read_json(link)
    j=df.to_json()
    j=loads(j)
    geo=pd.DataFrame.from_records(j['geolocation']).T.reset_index()
    data=pd.concat([df[['name','id','nametype','recclass','mass','year','reclat','reclong']],geo[['coordinates']]],axis=1)
    data.to_csv(f'{to_file}.csv',index=False)
    return f'saved to {to_file}.csv'
  except:
    print('error')


nasa_json_data('https://data.nasa.gov/resource/y77d-th95.json','nasa_data')

```
